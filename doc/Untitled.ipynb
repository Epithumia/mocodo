# Transformation en associations binaires

## Non-DF

```python
%%mocodo --mld
Varius: gravida, lorem
Euis, 1N Pharetra, 0N Varius, 1N Cras: praesent
Cras: lectus

:
Pharetra: augue dolor, nonummy
:
```

```python
%%mocodo --mld
Varius: gravida, lorem
DF1, 0N Varius, _11 Euis
Euis: _praesent
DF2, 1N Cras, _11 Euis
Cras: lectus

:
DF3, 1N Pharetra, _11 Euis
:
    
:
Pharetra: augue dolor, nonummy
:
```

## DF

```python
%%mocodo --mld
Varius: gravida, lorem
Euis, 11 Pharetra, 0N Varius, 1N Cras: praesent
Cras: lectus, cassis

:
Pharetra: augue dolor, nonummy
:
```

```python
%%mocodo --mld
Varius: gravida, lorem
:
Cras: lectus, cassis

DF1, 0N Varius, 11  Pharetra
Pharetra: augue dolor, nonummy, praesent
DF2, 1N Cras, 11 Pharetra
```

```python
%%mocodo --mld
Varius: gravida, lorem
DF1, 0N Varius, _11 Euis
Euis: _praesent
DF2, 1N Cras, _11 Euis
Cras: lectus, cassis

:
DF3, 11 Pharetra, 01 Euis
:
    
:
Pharetra: augue dolor, nonummy
:
```

```python
%%mocodo --mld
Varius: gravida, lorem
Euis, 0N Varius, 1N Cras: praesent
Cras: lectus, cassis

DF, 11 Pharetra, 0N Varius, 1N Cras
    
:
Pharetra: augue dolor, nonummy
:
```
